```python
x = [3, -2]
y = [2, -3]
```

```python
m = (y[1] - y[0]) / (x[1] - x[0])
x1=-1
y1=-2
```

```python
plt.plot(x_m,y_m, linestyle="none", marker="o", color="red", label="Manzanas")
plt.plot(x_p,y_p, linestyle="none", marker="o", color="red")
plt.plot(a_x,a_y, linestyle="none", marker="*", color="green", label="Peras")
plt.plot(x, y, marker="p")
plt.legend(loc=6)
plt.grid()
plt.show()
```

```python
m = (y[1] - y[0]) / (x[1] - x[0])
def funcion_w(m):
    return (m*1)
b = m*(-x1)+y1
w = [funcion_w(m),-1]

def funcion_peso(p1,p2,x1,y1,c):
    return (p1*x1)+(p2*y1)+b

pareja_x = [0, -2, -2, -2, 2, 0, 2, 2]
pareja_y = [2, 2, 2, -2, 2, -2, 0, -2]

#Ns neto de la recta
n1 = funcion_peso(w[0],w[1],pareja_x[0],pareja_y[0],b)
n2 = funcion_peso(w[0],w[1],pareja_x[1],pareja_y[1],b)
n3 = funcion_peso(w[0],w[1],pareja_x[2],pareja_y[2],b)
n4 = funcion_peso(w[0],w[1],pareja_x[3],pareja_y[3],b)
n5 = funcion_peso(w[0],w[1],pareja_x[4],pareja_y[4],b)
n6 = funcion_peso(w[0],w[1],pareja_x[5],pareja_y[5],b)
n7 = funcion_peso(w[0],w[1],pareja_x[6],pareja_y[6],b)
n8 = funcion_peso(w[0],w[1],pareja_x[7],pareja_y[7],b)
```

```python
#Funcion de activación
def fucion_activacion(v):
    if v < 0:
        return -1
    else:
        return 1
    
#Activacion para los Ns de la recta (1)
a1 = fucion_activacion(n1)
a2 = fucion_activacion(n2)
a3 = fucion_activacion(n3)
a4 = fucion_activacion(n4)
a5 = fucion_activacion(n5)
a6 = fucion_activacion(n6)
a7 = fucion_activacion(n7)
a8 = fucion_activacion(n8)


#Vector de salidas
salidas = [1, 1, 1, 1, 1, 0, 0, 0]

#Calcular el error recta
e1 = salidas[0]-a1
e2 = salidas[1]-a2
e3 = salidas[2]-a3
e4 = salidas[3]-a4
e5 = salidas[4]-a5
e6 = salidas[5]-a6
e7 = salidas[6]-a7
e8 = salidas[7]-a8
```
